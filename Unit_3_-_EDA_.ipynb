#### Fill in blanks with a brief description of each plot type:

  
Histogram: a bar graph that represents the frequency distribution of continuous data.

Density Plot / KDE: a smoothed representation of the distribution of data. It shows the estimated probability density function of the given random variable.

Bar Chart: a graph that represents categorical data using rectangular bars. The height of the bar represents the frequency or count of the data points in that category.

Line Plot: a graph that displays data along a number line by connecting data points with a line. It is used to show trends over time or the relationship between two variables.

Univariate Scatter Plot: a scatter plot that displays the relationship between one independent variable and one dependent variable. It is used to identify patterns and outliers in the data.

Violin Plot: a combination of a box plot and a density plot that shows the distribution of data. It is used to represent the probability density of the data at different values.

Box Plot: a graph that shows the summary of the set of data values having properties like minimum, first quartile, median, third quartile, and maximum.

Pie Chart: a circular graph that shows the proportions of different categories of data. It is used to represent the proportion of each category relative to the whole.

Rug Plot: a plot that displays individual data points along a number line. It is used to display the density of observations in a given dataset.

#### Crosstab of average salary grouped by education and industry

```python
# use pd.crosstab() to calculate mean salary in each industry grouped by education level
heat = pd.crosstab(index=salaryNum['industry'], columns=salaryNum['education'], values=salaryNum['salary'], aggfunc=np.mean).round(0)
```
